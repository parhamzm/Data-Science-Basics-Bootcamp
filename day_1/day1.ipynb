### Control Flow: Loops & Conditions
#### Goal: Iterate and filter data


```python
ages = [25, 30, 35, 40, 22, 58]
print("Ages over 30:")
for age in ages:
    if age > 30:
        print(age)
```

### Hands-On Challenge
Write a function to count unique values in a list

```python
def count_unique(items):
    result = {}
    for item in items:
        result[item] = result.get(item, 0) + 1
    return result

genders = ["Male", "Female", "Male"]
print(count_unique(genders))
```

## 3. Pandas Basics & Data Intuition (30 mins)
Why use Pandas instead of raw Python?

- Easier filtering, sorting, and analyzing
- Clean tabular structure (like Excel in code)


```python
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

### Key Methods for First Look
- `.head()` → First 5 rows
- `.info()` → Data types and missing values
- `.describe()` → Summary stats


```python
df.info()
df.describe()
```

### Filtering & Grouping in Pandas


```python
df[df['Age'] > 30].head()

# Grouping example
df.groupby('Pclass')['Age'].mean()
```

## 4. Mini-Project: Titanic Analysis (30 mins)
**Task**: Use Pandas to explore real data


```python
# Average age of survivors
df[df['Survived'] == 1]['Age'].mean()

# Compare survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

```python
# Visualization (optional if time)
df['Survived'].value_counts().plot(kind='bar', title='Survivors vs Non-Survivors')
```
